```python
Index = list(set(list(bme.index) + list(vh1.index )+ list(vh2.index) + list(sze.index) +list(lin.index) + list(jin.index )+ list(eqi.index) + list(syd.index)))
Index.sort()

df3 = pd.DataFrame({'linx': [lin.loc[Date, 'Per_8'] if Date in lin.index else np.nan for Date in Index],
                    'jinx': [jin.loc[Date, 'Per_8'] if Date in jin.index else np.nan for Date in Index],
                    'eqix': [eqi.loc[Date, 'Per_8'] if Date in eqi.index else np.nan for Date in Index],
                    'sydney': [syd.loc[Date, 'Per_8'] if Date in syd.index else np.nan for Date in Index],
                    'bme': [bme.loc[Date, 'Per_8'] if Date in bme.index else np.nan for Date in Index],
                    'vh1': [vh1.loc[Date, 'Per_8'] if Date in vh1.index else np.nan for Date in Index],
                    'vh2': [vh2.loc[Date, 'Per_8'] if Date in vh2.index else np.nan for Date in Index],
                    'szeged': [sze.loc[Date, 'Per_8'] if Date in sze.index else np.nan for Date in Index],
                   },index = Index)
plt.rc('legend', fontsize=18)
plt.rc('axes', labelsize=18)
df3.plot(figsize=(10,10)).grid()

#resample('D').mean()
```
